```python
# Data Analysis
# Visualize the relationship between ethnicity and perception of AI

# Identify relevant columns for yes/no questions and open-ended questions
yes_no_columns = ['ALGFAIR_W99', 'DISCRIM1_a_W99', 'DISCRIM1_b_W99', 'DISCRIM1_c_W99', 'DISCRIM1_d_W99', 'DISCRIM1_e_W99', 'DISCRIM1_f_W99',
                  'POSNEGAI_a_W99', 'POSNEGAI_b_W99', 'POSNEGAI_c_W99', 'POSNEGAI_d_W99', 'POSNEGAI_e_W99', 'POSNEGAI_f_W99',
                  'POSNEGHE_a_W99', 'POSNEGHE_b_W99', 'POSNEGHE_c_W99', 'POSNEGHE_d_W99', 'POSNEGHE_e_W99', 'POSNEGHE_f_W99']

# Create subplots for each yes/no question
fig, axs = plt.subplots(len(yes_no_columns), 1, figsize=(12, 6*len(yes_no_columns)), sharex=True)
fig.suptitle('Responses to Yes/No Questions')

for i, column in enumerate(yes_no_columns):
    ax = axs[i]
    # Count the occurrences of each response
    counts = df[column].value_counts()
    # Plot the counts
    counts.plot(kind='bar', ax=ax)
    ax.set_ylabel('Count')
    ax.set_title(column)
    ax.grid(axis='y')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Summarize responses to open-ended questions
open_ended_columns = ['EXCITEOE_W99_OE1', 'EXCITEOE_W99_OE2', 'EXCITEOE_W99_OE3', 'CONCERNOE_W99_OE1', 'CONCERNOE_W99_OE2', 'CONCERNOE_W99_OE3']
for column in open_ended_columns:
    print(f"Open-ended question: {column}")
    print(df[column].value_counts())
    print()
```

```python
age_labels = ['18-29', '30-49', '50-64', '65+', 'Refused']

male_data = df[df.F_GENDER == "A man"].groupby('F_AGECAT').size().reindex(age_labels, fill_value=0).values
female_data = df[df.F_GENDER == "A woman"].groupby('F_AGECAT').size().reindex(age_labels, fill_value=0).values
other_data = df[df.F_GENDER == "In some other way"].groupby('F_AGECAT').size().reindex(age_labels, fill_value=0).values
ref_data = df[df.F_GENDER == "Refused"].groupby('F_AGECAT').size().reindex(age_labels, fill_value=0).values

fig, ax = plt.subplots()

bar_width = 0.2
x = np.arange(len(age_labels))

male_plt_position = x - bar_width * 1.5
female_plt_position = x - bar_width / 2
other_plt_position = x + bar_width / 2
ref_plt_position = x + bar_width * 1.5


ax.bar(male_plt_position, male_data,width=bar_width,label='Male',color='blue')
ax.bar(female_plt_position, female_data,width=bar_width,label='Female',color='pink')
ax.bar(other_plt_position, other_data,width=bar_width,label='Other',color='yellow')
ax.bar(ref_plt_position, ref_data,width=bar_width,label='Refused',color='red')
plt.xticks(x)
ax.set_xticklabels(age_labels)
ax.set_ylabel("Count")
ax.set_xlabel("Age Group")
ax.set_title("Data by Age and Gender",fontsize=14)
plt.legend(loc='best')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1]) 
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3]) 
plt.show()
```
