```python
# Normal Distribution -- bell curve
## characterised by mean, std deviation, skewness, kurtosis
## continuous : -inf to +inf
```

```python
# Normal Distribution
np.random.seed(0)
data = np.random.normal(loc=50000, scale=10000, size=1000)
```

```python
sns.histplot(data, kde = True)

# x = salaries
# y = no. of people
```

```python
data.mean()
```

```python
data.std()
```

```python
# fit the distribution to data
mu, std = stats.norm.fit(data)
```

```python
mu
```

```python
std
```

```python
# QQ PLOT
stats.probplot(data, dist = 'norm', plot = plt)

## are the salaries fit in the same distr as theoritical dist.
```

```python
# MAKE PREDICTIONS:

#prob salary is between 40,000 to 60,000
stats.norm.cdf(60000, mu, std) - stats.norm.cdf(40000, mu, std)
```

```python
# Statistical Testing
# parametric - only for normally dist data
# proportions - beta 
# not normal - non parametric
```

```python
# statistical testing
from scipy.stats import ttest_ind, mannwhitneyu
```

```python
group_a = np.random.normal(30000, 5000, 1000)
group_b = np.random.normal(35000, 5000, 1000)
```

```python
t_stat, p_value = ttest_ind(group_a, group_b)
```

```python
print(p_value)  # less than .05 -- statistically significance
```

```python
# User Level Data
users = df.groupby('Customer_ID')[['Campaign_Exposure', 'Purchase_Amount', 'Premium_Product']].agg({'Campaign_Exposure':'max',
                                                                                                     'Purchase_Amount':'mean', 'Premium_Product':['mean','count']})
```

```python
# divide the dataframe based on campaign exposure
exp = users[users['Campaign_Exposure']['max']==1]
not_exp = users[users['Campaign_Exposure']['max']==0]
```

```python
# purchase amount
plt.hist(exp['Purchase_Amount'], bins=30)
```

```python
# purchase amount
plt.hist(not_exp['Purchase_Amount'], bins=30)
```

```python
t_stat, p_value = ttest_ind(not_exp['Purchase_Amount'], exp['Purchase_Amount'])
```

```python
print(p_value)
```

```python
t_stat, p_value = ttest_ind(not_exp['Purchase_Amount'], exp['Purchase_Amount'])

#non_exp have less
```

```python
print(t_stat)
print(p_value)
```

```python
not_exp['Purchase_Amount'].mean()
```

```python
exp['Purchase_Amount'].mean()
```

```python
#calculate percentage lift between test and control groups
(exp['Purchase_Amount'].mean()/not_exp['Purchase_Amount'].mean())-1

# 18% lift after applying the campaign
```

```python
# Non Parametric

u_stat, p_value = mannwhitneyu(not_exp['Premium_Product']['count'], exp['Premium_Product']['count'], alternative='less')
```

```python
print(p_value)
print(u_stat)
```

```python
u_stat, p_value = mannwhitneyu(not_exp['Premium_Product']['mean'], exp['Premium_Product']['mean'], alternative='less')
```

```python
print(p_value)
print(u_stat)
```
